### Ouliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Dikarenakan presentase yang tidak terlalu tinggi namun merupakan variable yang penting, maka dilakukan imputasi dengan metode yang paling sesuai, yaitu Winsorizing karena tetap mempertahankan struktur data tanpa membuang nilai ekstrem

```python
columns_to_impute = ["Sleep_Duration", "Sleep_Quality", "Screen_Time", "Caffeine_Intake", "Work_Hours", "Social_Interactions", "Cholesterol_Level", "Blood_Sugar_Level"]

for col in columns_to_impute:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menggunakan .loc[] agar tidak muncul SettingWithCopyWarning
    df.loc[:, col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Melakukan Pengecekan Ulang

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## Construct Data

```python
df['Wake_Up_Time'] = pd.to_datetime(df['Wake_Up_Time'], format='%I:%M %p', errors='coerce')
df['Bed_Time'] = pd.to_datetime(df['Bed_Time'], format='%I:%M %p', errors='coerce')

df['Time_in_Bed'] = (df['Wake_Up_Time'] + pd.Timedelta(days=1) - df['Bed_Time']).dt.total_seconds() / 3600

df['Sleep_Efficiency'] = df['Sleep_Duration'] / df['Time_in_Bed']

df['Wake_Up_Time'] = df['Wake_Up_Time'].dt.strftime('%H:%M')
df['Bed_Time'] = df['Bed_Time'].dt.strftime('%H:%M')

def calculate_health_risk(row):
    if (row['Blood_Pressure'] < 120 and
        row['Cholesterol_Level'] < 200 and
        row['Blood_Sugar_Level'] < 100):
        return 'Low Risk'
    elif (row['Blood_Pressure'] >= 120 or
          row['Cholesterol_Level'] >= 200 or
          row['Blood_Sugar_Level'] >= 100):
        return 'High Risk'
    else:
        return 'Unknown'

df['Health_Risk'] = df.apply(calculate_health_risk, axis=1)
```

## Korelasi Antar Variabel Numerik

```python
gender_mapping = {'Male': 0, 'Female': 1}
df['Gender_Num'] = df['Gender'].map(gender_mapping)
```

```python
risk_mapping = {'Low Risk': 0, 'High Risk': 1}
df['Health_Risk_Num'] = df['Health_Risk'].map(risk_mapping)
```

```python
stress_mapping = {'Low': 0, 'Medium': 1, 'High': 2}
df['Stress_Detection_Num'] = df['Stress_Detection'].map(stress_mapping)
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Age', 'Gender_Num','Sleep_Duration','Blood_Pressure', 'Cholesterol_Level','Blood_Sugar_Level', 'Sleep_Efficiency','Health_Risk_Num','Stress_Detection_Num']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```

Heatmap ini menampilkan korelasi antar variable yang menjadi faktor-faktor timbul atau meningkatkatnya tingkat stres. Dari heatmap ini ditemukan bahwa:
- Blood Pressure (0.50), Cholesterol Level (0.54), dan Blood Sugar Level (0.50) memiliki korelasi positif sedang terhadap Stress Detection. Ini menunjukkan bahwa individu dengan tekanan darah, kadar kolesterol, dan gula darah yang lebih tinggi cenderung memiliki tingkat stres yang lebih tinggi.
- Blood Pressure, Cholesterol Level, dan Blood Sugar Level juga memiliki korelasi yang sangat kuat di antara mereka (> 0.83). Ini mengindikasikan bahwa kenaikan satu variabel cenderung diikuti oleh kenaikan variabel lainnya
- Sleep Duration (-0.28) memiliki korelasi negatif lemah terhadap tingkat stres. Artinya, semakin sedikit durasi tidur, semakin tinggi potensi stres yang terdeteksi.
- Age memiliki korelasi positif sedang terhadap Blood Pressure (0.40), Cholesterol Level (0.47), dan Blood Sugar Level (0.50), menunjukkan bahwa usia yang lebih tua cenderung memiliki risiko kesehatan fisik yang lebih tinggi.
- Korelasi antara Age dan Stress Detection (0.28) menunjukkan bahwa semakin tua usia seseorang, cenderung ada sedikit peningkatan dalam tingkat stres.
- Sleep Efficiency memiliki korelasi negatif dengan Blood Pressure (-0.40), yang berarti semakin efisien tidur seseorang, semakin rendah tekanan darah mereka.
- Tidak ada hubungan kuat antara Sleep Efficiency dan Stress Detection (-0.10), yang berarti efisiensi tidur tidak terlalu memengaruhi deteksi stres secara signifikan.

Berdasarkan insight tersebut, organisasi atau individu dapat melakukan beberapa hal untuk mengurangi dan mencegah stres seperti:
- Meningkatkan kesadaran terhadap pengelolaan tekanan darah, kadar kolesterol, dan gula darah melalui pola makan sehat, olahraga rutin, dan pemantauan kesehatan secara berkala untuk mengurangi risiko stres yang lebih tinggi.
- Menerapkan konsistensi terhadap pola jam tidur yang ideal.
- Meningkatkan kewaspadaan kepada kelompok usia lanjut untuk melakukan pemeriksaan rutin untuk tekanan darah dan gula darah, karena memiliki korelasi dengan stres.
